```python
import pandas as pd
```

```python
df =  pd.read_csv('explanations_digits.csv')
```

```python
len_milp_relaxed = df['len_milp_relaxed']
```

```python
# transformar len_milp_relaxed em uma lista de listas
len_milp_relaxed = [ eval(str(x)) for x in len_milp_relaxed]
```

```python
import numpy as np
import matplotlib.pyplot as plt
array = np.array(len_milp_relaxed)
```

```python
array.shape
# calcular a média de cada coluna
array_mean = np.mean(array, axis=0)
array_mean
```

```python
plt.plot( [i/100 for i in range(1, 11, 1)], array_mean )
plt.xlabel('Delta')
plt.ylabel('Tamanho da explicação')
plt.title('Tamanho médio da explicação para cada delta')
plt.show()
```
