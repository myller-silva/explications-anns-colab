```python
import pandas as pd
```

```python
base_path = '.'
iris = pd.read_csv(f'{base_path}/explanations_iris.csv')
wine = pd.read_csv(f'{base_path}/explanations_wine.csv')
breast_cancer = pd.read_csv(f'{base_path}/explanations_breast_cancer.csv')
glass = pd.read_csv(f'{base_path}/explanations_glass.csv')
digits = pd.read_csv(f'{base_path}/explanations_digits.csv')


data = pd.concat([iris, wine, breast_cancer, glass, digits], ignore_index=True)
```

```python
import json
def get_last(x):
    return json.loads(x)[-1]

data["len_relaxed_global"] = data["len_milp_relaxed"].apply(lambda x: get_last(x))
data["time_relaxed_global"] = data["time_milp_relaxed"].apply(lambda x: get_last(x))
```

```python
desvio_padrao = (
    data.groupby(["dataset", "modelo"])[["len_milp", "len_relaxed_global", "time_milp", "time_relaxed_global"]]
    .std()
    .reset_index()
)
desvio_padrao
```

```python
media = data.groupby(["dataset", "modelo"])[["len_milp", "len_relaxed_global", "time_milp", "time_relaxed_global"]].mean().reset_index()
media
```

```python
# renomear os modelos para facilitar a visualização
data["modelo"] = data["modelo"].replace({
    "model_1layers_20neurons.h5": "1L20N",
    "model_2layers_20neurons.h5": "2L20N",
    "model_3layers_20neurons.h5": "3L20N",
    
})

grouped = data.groupby(["dataset", "modelo"])

# Calcular a média e o desvio padrão
mean_std = grouped[["len_relaxed_global", "time_relaxed_global"]].agg(['mean', 'std']).reset_index()

# Exibir como tabela
mean_std.columns = ['dataset', 'modelo', 'len_relaxed_global_mean', 'len_relaxed_global_std', 'time_relaxed_global_mean', 'time_relaxed_global_std']
mean_std["len_mean_std"] = mean_std.apply(lambda x: f'{x["len_relaxed_global_mean"]:.2f} ± {x["len_relaxed_global_std"]:.2f}', axis=1)
mean_std["time_mean_std"] = mean_std.apply(lambda x: f'{x["time_relaxed_global_mean"]:.2f} ± {x["time_relaxed_global_std"]:.2f}', axis=1)

# Exibir como tabela
print("Tabela 1: Média e desvio padrão do tamanho e tempo de execução das explicações relaxadas")
mean_std[["dataset", "modelo", "len_mean_std", "time_mean_std"]]
```
