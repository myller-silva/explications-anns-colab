```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, scale
from sklearn.model_selection import train_test_split
```

```python

dataset = pd.read_csv('glass.tsv', sep='\t')
target_map = {1:0, 2:1, 3:2, 5:3, 7:4}
dataset['target'] = dataset['target'].apply(lambda x: target_map[x])
data, target = dataset.drop('target', axis=1), dataset['target']
# target
```

```python
x_train, x_test, y_train, y_test = train_test_split(data, target, test_size=0.2, random_state=42)
```

```python
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

```python
# Normalizando os dados
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

```python
import numpy as np

# Concatenate the datasets
y_train = y_train.to_numpy()
y_test = y_test.to_numpy()

data_train = np.concatenate((x_train, y_train.reshape(-1, 1)), axis=1)
data_test = np.concatenate((x_test, y_test.reshape(-1, 1)), axis=1)
```
